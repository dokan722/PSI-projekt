```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('anime.csv', index_col=0)
print(data.shape)
data.head()
```

```python
data.info()
```

```python
data = data.drop(['Name', 'Score', 'English name', 'Japanese name', 'Aired', 'Ranked', 'Popularity'], axis=1)
data = data.iloc[: , :-16]
print(data.shape)
data.head()
```

```python
import itertools
import collections
import matplotlib.pyplot as plt

data['Genres'] = data['Genres'].map(lambda s: s.split(', '))
genres = data['Genres'].tolist()
flatted = list(itertools.chain(*genres))
counted = collections.Counter(flatted).most_common()

genres = list(list(zip(*counted))[0])
num = list(list(zip(*counted))[1])

revg = genres[::-1]
revn = num[::-1]
y = np.arange(len(revg))

print(counted)
plt.figure(figsize=(10, 35))
plt.barh(y, revn, height=0.6)
plt.yticks(y, revg)
plt.show()
```

```python
def main_genre(row):
    genre_list = row['Genres']
    for gen in genres:
        if gen in genre_list:
            return gen
    return genres[0]

data['Genres'] = data.apply(lambda row: main_genre(row), axis=1)
data
```

```python
print(data['Genres'].value_counts())
```

```python
print(data['Type'].value_counts())
```

```python
def map_episodes(row):
    num = row['Episodes']
    num = int(num) if num != 'Unknown'  else 1
    if num == 1:
        return 1
    elif num < 20:
        return 10
    elif num < 30:
        return 20
    elif num < 100:
        return 50
    return 100

data['Episodes'] = data.apply(lambda row: map_episodes(row), axis=1)
data
```

```python
print(data['Episodes'].value_counts())
```

```python
print(data['Premiered'].value_counts())
```

```python
data = data.drop(["Premiered"], axis=1)
data
```

```python
data['Producers'] = data['Producers'].map(lambda s: s.split(', '))
producers = data['Producers'].tolist()
flatted = list(itertools.chain(*producers))
counted = collections.Counter(flatted).most_common()

producers = list(list(zip(*counted))[0])

def main_producer(row):
    prod_list = row['Producers']
    for prod in producers:
        if prod in prod_list:
            return prod
    return producers[0]

data['Producers'] = data.apply(lambda row: main_producer(row), axis=1)
data
```

```python
print(data['Producers'].value_counts())
```

```python
data['Licensors'] = data['Licensors'].map(lambda s: s.split(', '))
licensors = data['Licensors'].tolist()
flatted = list(itertools.chain(*licensors))
counted = collections.Counter(flatted).most_common()

licensors = list(list(zip(*counted))[0])

def main_licensor(row):
    prod_list = row['Licensors']
    for lic in licensors:
        if lic in prod_list:
            return lic
    return licensors[0]

data['Licensors'] = data.apply(lambda row: main_licensor(row), axis=1)
data
```

```python
print(data['Licensors'].value_counts())
```

```python
data['Studios'] = data['Studios'].map(lambda s: s.split(', '))
studios = data['Studios'].tolist()
flatted = list(itertools.chain(*studios))
counted = collections.Counter(flatted).most_common()

studios = list(list(zip(*counted))[0])

def main_studio(row):
    studio_list = row['Studios']
    for studio in studios:
        if studio in studio_list:
            return studio
    return studios[0]

data['Studios'] = data.apply(lambda row: main_studio(row), axis=1)
data
```

```python
def map_duration(row):
    splitted = row['Duration'].split(" ")
    if len(splitted) < 2:
        return 'Unknown'
    if splitted[1] == 'sec.':
        return 'short'
    elif splitted[1] == 'min.':
        if int(splitted[0]) < 10:
            return 'short'
        elif int(splitted[0]) < 30:
            return 'episode'
        else:
            return 'long episode'
    elif splitted[1] == 'hr.':
        return 'film'
    else:
        return 'Unknown'


data['Duration'] = data.apply(lambda row: map_duration(row), axis=1)
data
```

```python
print(data['Duration'].value_counts())
```

```python
print(data['Rating'].value_counts())
```

```python
data['Members'].describe().apply("{0:.3f}".format)
```

```python
def map_members(row):
    num = int(row['Members'])
    if num <= 336:
        return 0
    elif num <= 13223:
        return 1
    else:
        return 2


data['Members'] = data.apply(lambda row: map_members(row), axis=1)
data
```

```python
print(data['Members'].value_counts())
```

```python
from sklearn import preprocessing

tables_to_transform = ['Genres', 'Type', 'Producers', 'Licensors', 'Studios', 'Source', 'Duration', 'Rating']

for label in tables_to_transform:
    le = preprocessing.LabelEncoder()
    le.fit(data[label])
    data[label] = le.transform(data[label])
```

```python
import seaborn as sns

sns.set(style='darkgrid')

corr = data.corr()
f, ax = plt.subplots(figsize=(25, 25))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, cmap=cmap, vmin=-1, vmax=1,
            square=True, xticklabels=True, yticklabels=True,
            linewidths=.5, cbar_kws={'shrink': .5}, ax=ax)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = data.loc[:, data.columns != 'Members']
y = data['Members']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

kfold = model_selection.KFold(n_splits=5, random_state=None)

import warnings
warnings.filterwarnings('ignore')
```

1. Logistic Regression

```python
pipe = Pipeline([
    ('preprocessing', StandardScaler()),
    ('classifier', LogisticRegression(class_weight='balanced'))])

param_grid = {
            'preprocessing': [StandardScaler(), None],
            'classifier__C': [0.01, 0.1, 1, 10],
}

lr = GridSearchCV(pipe, param_grid, cv=kfold, return_train_score=True, scoring='f1')

lr.fit(X_train, y_train)
lr.best_params_
```

2. SVM linear

```python
pipe = Pipeline([
    ('preprocessing', StandardScaler()),
    ('classifier', SVC(kernel='linear', probability=True, verbose=True))])

param_grid = {
            'classifier__C': [0.01, 0.1, 1]
}

svm_linear = GridSearchCV(pipe, param_grid, cv=kfold, error_score='raise', n_jobs=-1, verbose=2)

svm_linear.fit(X_train, y_train)
svm_linear.best_params_
```

14. Keras

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import History
import tensorflow as tf

history = History()
keras = Sequential()
keras.add(Dense(200,activation="sigmoid", input_shape=(X_train.shape[1],)))
keras.add(Dense(100,activation="sigmoid"))
keras.add(Dense(50,activation="sigmoid"))
keras.add(Dense(10,activation="sigmoid"))
keras.add(Dense(3,activation="softmax"))
keras.summary()
```

```python
keras.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy', tf.keras.metrics.Recall(), tf.keras.metrics.Precision(), tf.keras.metrics.AUC()])
```

```python
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=50, mode='min', verbose=1)

history = keras.fit(X_train, tf.keras.utils.to_categorical(y_train), validation_data=(X_test, tf.keras.utils.to_categorical(y_test)),
                    batch_size=64, epochs=500, callbacks=[early_stopping])
```

```python
pd.DataFrame(history.history).plot(figsize=(24, 15))
plt.grid(True)
plt.gca().set_ylim(0, 1)
plt.show()
```

```python
from sklearn import  metrics
from sklearn.metrics import confusion_matrix, roc_curve, auc


models = []
models.append(('LR', lr.best_estimator_))
models.append(('SVM linear', svm_linear.best_estimator_))
models.append(('SVM rbf', svm_rbf.best_estimator_))
models.append(('SVM poly', svm_poly.best_estimator_))
models.append(('SVM sigmoid', svm_sigmoid.best_estimator_))
models.append(('KNeighborsClassifier', kneighbors.best_estimator_))
models.append(('DecisionTreeClassifier', decisiontree.best_estimator_))
models.append(('RandomForestClassifier', randomforest.best_estimator_))
models.append(('BaggingClassifier', bagging.best_estimator_))
models.append(('ExtraTreesClassifier', extratrees.best_estimator_))
models.append(('AdaBoostClassifier', adaboost.best_estimator_))
models.append(('GradientBoostingClassifier', gradientbost.best_estimator_))
models.append(('MLPClassifier', mlp.best_estimator_))
#models.append(('Keras', keras))



precision_score = []
recall_score = []
f1_score = []
accuracy_score = []
roc_auc_score = []
for name, model in models:
    pred = model.predict(X_test)
    pred_prob = model.predict_proba(X_test)
    precision_score.append(metrics.precision_score(y_test, pred,average='macro'))
    recall_score.append(metrics.recall_score(y_test, pred,average='macro'))
    f1_score.append( metrics.f1_score(y_test, pred,average='macro'))
    accuracy_score.append(metrics.accuracy_score(y_test, pred))
    roc_auc_score.append(metrics.roc_auc_score(y_test, pred_prob, multi_class='ovr'))
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 6))
    fig.suptitle(name, fontsize=25)
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt='g', cmap="RdPu", ax=ax1)
    for (idx, c_label) in enumerate(['Not popular', 'Normal', 'Popular']):
        fpr, tpr, thresholds = roc_curve(y_test, pred_prob[:, idx], pos_label=idx)
        ax2.plot(fpr, tpr)
    ax2.plot(fpr, fpr, '--', color = '0.75')
    ax2.legend(['Not popular', 'Normal', 'Popular', 'Random guessing'])
```

```python
d = {'precision_score': precision_score,
     'recall_score': recall_score,
     'f1_score': f1_score,
     'accuracy_score' : accuracy_score,
     'roc_auc' : roc_auc_score
    }
df = pd.DataFrame(data=d)
df.insert(loc=0, column='Model', value=['LR', 'SVM linear', 'SVC rbf', 'SVM poly', 'SVM sigmoid', 'KNeighborsClassifier',
                                        'DecisionTreeClassifier', 'RandomForestClassifier', 'BaggingClassifier', 'ExtraTreesClassifier',
                                        'AdaBoostClassifier', 'GradientBoostingClassifier', 'MLPClassifier'])
df
```
